```python
import numpy as np
import matplotlib.pyplot as plt

#Supongamos que los leads aumentan por campaña.Pero tambien se pierden por
#bloqueos de Meta o por desinteres o ..

#Ecuación: dL/dt = k * L
# L = número de leads
#k < 0 representa la perdida

k = -0.3  # perdida de leads (bloqueos, rechazo)



def f(t, L):
    return k * L

def euler_method(f, t0, L0, t_end, h):
    t_values = [t0]
    L_values = [L0]

    t = t0
    L = L0

    while t < t_end:
        L = L + h * f(t, L)
        t = t + h
        t_values.append(t)
        L_values.append(L)

    return t_values, L_values

# serian paramtros

t0 = 0
L0 = 100
t_end = 1      #intervalos [0,1]
h = 0.1        #paso h = 0.2

#respuesta euler 
t_euler, L_euler = euler_method(f, t0, L0, t_end, h)

#Separaci'on de variables
t_exact = np.linspace(t0, t_end, 100)
L_exact = L0 * np.exp(k * t_exact)


#grafica

plt.plot(t_euler, L_euler, 'o-', label="Método de Euler")
plt.plot(t_exact, L_exact, '-', label="Solución Analítica")
plt.xlabel("Tiempo")
plt.ylabel("Número de leads")
plt.title("Comparación: Solución Analítica vs Método de Euler")
plt.grid(True)
plt.legend()
plt.show()

#El metodo Euler aproxima la soluci'on  analitica
# la diferencia depende del tamaño del paso h

```
