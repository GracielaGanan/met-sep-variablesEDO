```python
import numpy as np
import matplotlib.pyplot as plt

#Supongamos que los leads aumentan por campaña.Pero tambien se pierden por
#bloqueos de Meta o por desinteres o ..

#Ecuación: dL/dt = k * L
# L = número de leads
k = -0.3  # perdida de leads (bloqueos, rechazo)



def f(t, L):
    return k * L

def euler_method(f, t0, L0, t_end, h):
    t_values = [t0]
    L_values = [L0]

    t = t0
    L = L0

    while t < t_end:
        L = L + h * f(t, L)
        t = t + h
        t_values.append(t)
        L_values.append(L)

    return t_values, L_values

# serian paramtros

t0 = 0
L0 = 100
t_end = 2
h = 0.1

t_values, L_values = euler_method(f, t0, L0, t_end, h)

#grafica

plt.plot(t_values, L_values, label="Leads (Euler)")
plt.xlabel("Tiempo")
plt.ylabel("Número de leads")
plt.title("Simulación de leads con Método de Euler")
plt.grid(True)
plt.legend()
plt.show()

# Este modelo simula la evolución del número de leads en una campaña de Hybridge
#debido a bloqueos o pérdida de interés.
#El método de Euler permite aproximar la solución paso a paso.

```
